```python
%autoreload

import jax
import jax.numpy as jnp
import matplotlib
import matplotlib.pyplot as plt
import numpyro
import numpyro.distributions as dist
import pears
import seaborn as sns
from jax import random
from jax.flatten_util import ravel_pytree
from jax.scipy.special import logsumexp
from numpyro import optim
from numpyro.diagnostics import print_summary
from numpyro.distributions import constraints
from numpyro.infer import MCMC, NUTS, SVI, Predictive, Trace_ELBO
from numpyro.infer.autoguide import AutoBNAFNormal, AutoIAFNormal, AutoNormal
from numpyro.infer.reparam import NeuTraReparam
```

```python
def run_neutra(
    model,
    data,
    key=random.PRNGKey(0),
    hidden_factors=8,
    opt=optim.Adam(1e-2),
    num_iters=10_000,
    num_samples=1_000,
    num_warmup=2_000,
    num_chains=1,
):
    keys = random.split(key)
    guide = AutoBNAFNormal(model, hidden_factors=[hidden_factor, hidden_factor])
    svi = SVI(model, guide, opt, Trace_ELBO())
    svi_result = svi.run(keys[0], num_iters, data)
    guide_samples = guide.sample_posterior(
        keys[1], svi_result.params, sample_shape=(num_samples,)
    )
    neutra = NeuTraReparam(guide, svi_result.params)
    neutra_model = neutra.reparam(model)
    nuts_kernel = NUTS(neutra_model, dense_mass=True)
    mcmc = MCMC(
        nuts_kernel,
        num_warmup=num_warmup,
        num_samples=num_samples,
        num_chains=num_chains,
        progress_bar=True,
        jit_model_args=True,
    )
    mcmc.run(keys[2], data)
    zs = mcmc.get_samples(group_by_chain=True)
    print(zs.keys())
    samples = neutra.transform_sample(zs)
    print_summary(samples)
    return neutra, mcmc, guide_samples, samples
```

# Problem 1: radial velocities

```python
t = jnp.array([0.0, 49.452, 95.393, 127.587, 190.408])
rv = jnp.array([38.77, 39.70, 37.45, 38.31, 38.31])
err = jnp.array([0.184, 0.261, 0.112, 0.155, 0.223])
plt.errorbar(t, rv, yerr=err, ls="none", marker="o")
```

```python
@jax.jit
def forward(params, data):
    omega = 2.0 * jnp.pi / jnp.exp(params["log_P"])
    v = params["v0"] + params["K"] * jnp.cos(omega * data["t"] + params["phi"])
    return v
```

```python
def model(data=None):
    K = numpyro.sample(
        "K",
        dist.Normal(data["mu_K"], data["sigma_K"]),
    )
    v0 = numpyro.sample("v0", dist.Normal(0.0, data["sigma_v0"]))
    phi = numpyro.sample("phi", dist.Uniform(0.0, 2.0 * jnp.pi))
    log_P = numpyro.sample(
        "log_P", dist.Uniform(jnp.log(data["Pmin"]), jnp.log(data["Pmax"]))
    )
    v = forward(dict(log_P=log_P, v0=v0, K=K, phi=phi), data)
    numpyro.sample("obs", dist.Normal(v, data["err"]), obs=data["v"])
```

```python
data = dict(
    t=t,
    v=rv,
    err=jnp.array([1.8]),
    Pmin=jnp.exp(3),
    Pmax=jnp.exp(5),
    sigma_v0=jnp.array([5.0]),
    mu_K=jnp.array([5.0]),
    sigma_K=jnp.array([3.0]),
)
```

```python
neutra, mcmc, guide_samples, samples = run_neutra(model, data)
```

```python
pears.pears(jax.tree_map(jnp.ravel, samples));
```

```python
tnew = jnp.linspace(t.min(), t.max(), 500)
ppsim = jax.vmap(forward, in_axes=({k: 0 for k in samples.keys()}, None))(
    jax.tree_map(jnp.ravel, samples), dict(t=tnew)
)
```

```python
for i in range(200):
    plt.plot(tnew, ppsim[i], c="gray", alpha=0.3)
plt.errorbar(t, rv, yerr=err, ls="none", marker="o")
```

```python
def model(dim=10):
    y = numpyro.sample("y", dist.Normal(0, 3))
    numpyro.sample("x", dist.Normal(jnp.zeros(dim - 1), jnp.exp(y / 2)))
```

```python
guide = AutoNormal(model)
svi = SVI(model, guide, optim.Adam(1e-2), Trace_ELBO())
svi_result = svi.run(random.PRNGKey(0), 5_000)
```

```python
predictive = Predictive(guide, params=svi_result.params, num_samples=1000)
samples = predictive(random.PRNGKey(1))
```
