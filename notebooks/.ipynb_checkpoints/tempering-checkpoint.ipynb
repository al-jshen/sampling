```python
from functools import partial

import jax
import jax.numpy as jnp
import jax.random as random
import jax.scipy.special as special
import jax.scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from jax import lax
```

```python
@jax.jit
def logpdf(x):
    return special.logsumexp(
        jnp.array(
            [
                stats.norm.logpdf(x, -3.0, 0.25),
                stats.norm.logpdf(x, 0.0, 0.4),
                stats.norm.logpdf(x, 2.0, 0.1),
            ]
        ),
        axis=0,
    )
```

```python
xs = np.linspace(-5, 5)
ys = logpdf(xs)
for t in [np.sqrt(3) ** i for i in range(5)]:
    plt.plot(xs, jnp.exp(ys / t), label=t)
plt.legend()
```

```python
@jax.jit
def swapright_ix(x):
    ix = jnp.arange(x.size)
    ix1 = ix[::2]
    ix2 = ix1 + 1
    s = jnp.minimum(ix1.size, ix2.size) // 2 * 2
    ix1 = ix1[:s]
    ix2 = ix2[:s]
    return ix1, ix2
```

```python
@jax.jit
def swapright(x):
    ix1, ix2 = swapright_ix(x)
    temp = x[ix1]
    x = x.at[ix1].set(x[ix2])
    x = x.at[ix2].set(temp)
    return x
```

```python
@jax.jit
def swapleft_ix(x):
    ix = jnp.arange(x.size)
    ix1 = ix[::-2]
    ix2 = ix1 - 1
    s = jnp.minimum(ix1.size, ix2.size) // 2 * 2
    ix1 = ix1[:s]
    ix2 = ix2[:s]
    return ix1, ix2
```

```python
@jax.jit
def swapleft(x):
    ix1, ix2 = swapleft_ix(x)
    temp = x[ix1]
    x = x.at[ix1].set(x[ix2])
    x = x.at[ix2].set(temp)
    return x
```

```python
# @partial(jax.jit, static_argnums=(0,))
def rwmh(target, x, temps, stepsize, key, *args):
    n_temps, n_dims = x.shape
    temps = temps.reshape(n_temps, 1)
    keys = random.split(key, 4)

    # normal update, with temperatures
    x_new = random.normal(keys[0], shape=x.shape) * stepsize + x
    log_accept = target(x_new, *args) - target(x, *args)
    log_accept_with_temp = log_accept / temps
    u = jnp.log(random.uniform(keys[1], shape=log_accept.shape))
    new = jnp.where(u < log_accept_with_temp, x_new, x)
    # print(new)
    # coupling/swap update

    zero_one = random.randint(keys[2], shape=(), minval=0, maxval=2)  # 0 or 1
    ix_i = jnp.array([zero_one + 2 * i for i in range(n_temps // 2)])
    ix_j = ix_i + 1
    swap_ix = jnp.arange(n_temps)
    swap_ix = swap_ix.at[ix_i].set(ix_j)
    swap_ix = swap_ix.at[ix_j].set(ix_i)
    swapped = new[swap_ix]
    temps_swapped = temps[swap_ix]

    swap11 = target(new, *args) / temps
    swap22 = target(swapped, *args) / temps_swapped
    swap21 = target(swapped, *args) / temps
    swap12 = target(new, *args) / temps_swapped

    proposal = swap12 + swap21
    current = swap11 - swap22

    log_accept_with_swap = proposal - current
    u = jnp.log(random.uniform(keys[3], shape=log_accept_with_swap.shape))
    new_swapped = jnp.where(u < log_accept_with_swap, swapped, new)
    return new_swapped
```

```python
@partial(jax.jit, static_argnums=(0, 4))
def sample(target, x, stepsize, temps, steps, seed=0, *args):
    # tempered_target = make_tempered_fn(target)
    # assert temps.size % 2 == 1  # odd # of temps
    # assert jnp.all(jnp.diff(temps) < 0)  # temps are decreasing
    # assert temps[-1] == 1.0  # last temp is 1

    @jax.jit
    def inner_fn(carry, i):
        key, x = carry
        key, subkey = random.split(key)
        y = rwmh(target, x, temps, stepsize, subkey, *args)
        return (key, y), y

    key = random.PRNGKey(seed)
    _, stacked = lax.scan(inner_fn, (key, x), None, length=steps)

    return stacked
```

```python
dim = 1
chains = 4
nswap = 2
niter = 6000
nwarmup = 3000
nthin = 2
temps = jnp.array([2**i for i in range(9)])[::-1]
key = random.PRNGKey(0)
init = random.normal(key, shape=(chains, temps.size, dim))
# init = jnp.ones((chains, temps.size, dim))
```

```python
sample(logpdf, init[0], 0.1, temps, niter, 0).shape
```

```python
chain = jax.vmap(sample, in_axes=(None, 0, None, None, None, 0))(
    logpdf, init, 0.2, temps, niter, jnp.arange(chains)
)[:, nwarmup:, ...][:, ::nthin, ...]
```

```python
fig, ax = plt.subplots(2, temps.size, figsize=(temps.size * 4, 7))
for i in range(temps.size):
    sns.histplot(chain[:, :, i, 0].ravel(), stat="density", bins=50, ax=ax[0, i])
    ax[0, i].plot(xs, np.exp(logpdf(xs) / temps[i]), c="r")
    for c in range(chains):
        ax[1, i].plot(chain[c, :, i, 0], alpha=0.6)
```
